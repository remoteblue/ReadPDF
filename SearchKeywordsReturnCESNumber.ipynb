```python
#######   INPUT         #####################
# !!! All inputs should be provided here!!!

# Specify the path to your PDF file
pdf_file_path = '/Users/htan/Documents/Parenting/2024Summer/20231120_PPL_Kids_On_Campus_Schedule_Summer_2024_05.pdf'
# keywords of all courses
keywords = ['Dinosaur Detective', 'Junior Imagineer']
######    End of Input #######################

def pdf_to_text(pdf_file_path):
    # Open the PDF file
    doc = fitz.open(pdf_file_path)
    
    # Initialize an empty string for text content
    text_content = ''
    
    # Iterate through each page
    for page in doc:
        # Extract text from the current page
        text_content += page.get_text()
    
    # Close the document
    doc.close()
    
    return text_content


text = pdf_to_text(pdf_file_path)

# Optionally, print the text to the console
#print(text)
```
